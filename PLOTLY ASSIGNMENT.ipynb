Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three dimensional space

                np.random.seed(30) 
                data = { 'X': np.random.uniform(-10, 10, 300), 
                         'Y': np.random.uniform(-10, 10, 300), 
                         'Z': np.random.uniform(-10, 10, 300)}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X'], df['Y'], df['Z'])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import pandas as pd
import seabron as sns
import matplotlib.pyplot as plt

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data)

sns.violinplot(x='Grade', y='Score', data=df_grades)
plt.title('Distribution of Scores Across Grades')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

sns.set(rc={'figure.figsize': (10, 6)})
sns.heatmap(df_sales.pivot_table(index='Month', columns='Day', values='Sales'), annot=True, fmt='d')
plt.title('Sales Variation Across Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and

     'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

sns.heatmap(df.pivot_table(index='Month', columns='Day', values='Sales'), annot=True, fmt='d')

plt.title('Sales Variation Across Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')

plt.show()
```

Q4. Using the given x and y data, generate a 3D surface plot to visualize the function

                   x = np.linspace(-5, 5, 100) 
                   y = np.linspace(-5, 5, 100) 
                   x, y = np.meshgrid(x, y) 
                   z = np.sin(np.sqrt(x**2 + y**2))  
                    data = { 
                      'X': x.flatten(), 
                      'Y': y.flatten(), 
                      'Z': z.flatten() 
                                      } 
                   df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
z = np.sin(np.sqrt(X**2 + Y**2))

# Create a DataFrame
data = {
    'X': X.flatten(),
    'Y': Y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, z, cmap='viridis', alpha=0.7)

# Add contour lines for better visualization
ax.contourf(X, Y, z, zdir='z', offset=-1, cmap='viridis', alpha=0.3)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot of sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

Q5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

                    data = { 
                   'Country': ['USA', 'Canada', 'UK' ,'Germany', 'France'], 
                   'Population': 
                    np.random.randint(100, 1000, 5), 
                   'GDP': np.random.randint(500, 2000, 
                    5) 
                     }

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.7)

for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=10)

plt.title('Population vs GDP')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```
