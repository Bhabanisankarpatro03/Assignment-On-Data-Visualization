Creating a Scatter Plot with a Synthetic Dataset

Understanding the Task:

We'll generate a synthetic dataset with two variables, x and y, and then visualize their relationship using a scatter plot. This will allow us to explore different patterns and trends between the variables.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)  # Set a seed for reproducibility
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)  # Introduce some noise

plt.scatter(x, y)

plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.title("Scatter Plot of X and Y")

plt.show()
```

Explanation :

1. Import necessary libraries: matplotlib.pyplot for plotting and numpy for numerical operations.

2. Generate synthetic data:

*np.random.seed(42): Sets a seed for random number generation, ensuring reproducibility.

*x = np.random.randn(100): Creates an array of 100 random numbers from a standard normal distribution.

*y = 2 * x + np.random.randn(100): Generates a dependent variable y based on x with added noise. This simulates a linear relationship with some randomness.

3. Create the scatter plot:
   
*plt.scatter(x, y): Plots the x and y values as points on the scatter plot.

4.Add labels and a title:

*plt.xlabel("Variable X"), plt.ylabel("Variable Y"), and plt.title("Scatter Plot of X and Y") add labels and a title to the plot for clarity.

5. Show the plot:

*plt.show() displays the created scatter plot.

Customization:

You can customize the plot further by:

* Adjusting the number of data points: Change the 100 in np.random.randn(100) to generate more or fewer points.

* Modifying the relationship between x and y: Experiment with different coefficients in the y = 2 * x + np.random.randn(100) equation to create different patterns (e.g., quadratic, exponential).

* Adding color or size variations: Use the c and s parameters in plt.scatter to color-code or size the points based on additional variables.
Changing the plot style: Use plt.style.use() to apply different styles (e.g., plt.style.use('ggplot')).

Q2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable

## Generating and Visualizing a Random Dataset

**Understanding the Task:**
We'll generate a random dataset of numerical values and then visualize its distribution to understand its characteristics.

**Python Code:**

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Distribution of Random Numbers")

plt.show()
```

**Explanation:**

1. **Import necessary libraries:** `numpy` for numerical operations and `matplotlib.pyplot` for plotting.
2. **Generate a random dataset:**
   - `np.random.normal(loc=0, scale=1, size=1000)`: Generates 1000 random numbers from a normal distribution with a mean of 0 and a standard deviation of 1. You can adjust these parameters to create different distributions.
3. **Visualize the distribution using a histogram:**
   - `plt.hist(data, bins=30, density=True)`: Creates a histogram with 30 bins and normalizes the y-axis to represent density instead of frequency. This allows for easier comparison of distributions.
4. **Add labels and a title:**
   - `plt.xlabel("Value")`, `plt.ylabel("Density")`, and `plt.title("Distribution of Random Numbers")` add labels and a title to the plot for clarity.
5. **Show the plot:**
   - `plt.show()` displays the created histogram.

**Customization:**
- **Change the distribution:** Experiment with different distributions like `np.random.uniform` (uniform distribution), `np.random.poisson` (Poisson distribution), or `np.random.exponential` (exponential distribution) by replacing `np.random.normal`.
- **Adjust the number of bins:** Modify the `bins` parameter in `plt.hist` to control the number of bars in the histogram.
- **Add a probability density function (PDF):** Overlay the theoretical PDF of the chosen distribution on the histogram for comparison.
- **Explore other visualizations:** Consider using a kernel density estimate (KDE) plot or a box plot for different perspectives on the distribution.

By exploring these variations, you can visualize and analyze various random distributions and gain insights into their characteristics.


Q3. Create a dataset representing categories and their corresponding values. Compare different categories 
based on numerical values.

Creating a Dataset and Comparing Categories

Dataset Creation:

Let's create a dataset representing different fruits and their corresponding vitamin C content.

```python
import pandas as pd

data = {
    'Fruit': ['Apple', 'Orange', 'Banana', 'Strawberry', 'Kiwi'],
    'Vitamin C (mg)': [10, 50, 12, 60, 90]
}

df = pd.DataFrame(data)
print(df)
```

Comparison and Analysis:

We can use various methods to compare the categories (fruits) based on their numerical values (vitamin C content):

1.Descriptive Statistics:

* Calculate the mean, median, and mode of vitamin C content for each fruit.
* Find the maximum and minimum vitamin C values.
* Determine the standard deviation to measure the spread of values.

```python
print(df.describe())
```

2 Visualization:

* Create a bar chart to compare vitamin C content visually.
* Use a box plot to show the distribution of vitamin C content for each fruit.

```python
import matplotlib.pyplot as plt

plt.bar(df['Fruit'], df['Vitamin C (mg)'])
plt.xlabel('Fruit')
plt.ylabel('Vitamin C (mg)')
plt.title('Vitamin C Content in Fruits')
plt.show()
```

3 Statistical Tests:

* Perform t-tests or ANOVA to compare the means of vitamin C content between different fruits.

```python
from scipy.stats import ttest_ind

result = ttest_ind(df['Vitamin C (mg)'][df['Fruit'] == 'Apple'], df['Vitamin C (mg)'][df['Fruit'] == 'Orange'])
print(result)
```

Q4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical 
variable across different categories

Generating a Dataset with Categories and Numerical Values

Let's create a hypothetical dataset about the sales of different products in various regions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

regions = ["North", "South", "East", "West"]
products = ["Product A", "Product B", "Product C"]

np.random.seed(42)  # For reproducibility
sales_data = np.random.randint(100, 1000, size=(len(regions), len(products)))

df = pd.DataFrame(sales_data, index=regions, columns=products)

print(df)
```

Visualizing the Distribution of a Numerical Variable Across Different Categories

Let's visualize the distribution of sales for "Product A" across different regions.

```python
# Extract sales data for "Product A"
product_a_sales = df["Product A"]

# Create a box plot
plt.figure(figsize=(8, 5))
plt.boxplot([product_a_sales[region] for region in regions])
plt.title("Sales of Product A by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(range(1, len(regions) + 1), regions)
plt.show()
```
